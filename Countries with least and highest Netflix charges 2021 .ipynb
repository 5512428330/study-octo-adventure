```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the csv file with limit 20 rows
df = pd.read_csv('Netflix subscription fee Dec-2021.csv')
df.head(20)
```

```python
#Rearranging the dataframe in descending order by Cost Per Month Basic Plan
dataframe = df.iloc[:, :8].sort_values(by = 'Cost Per Month - Basic ($)', ascending = False)
dataframe
```

```python
#Finding the average value of all the three types of plan
df['Avg_Cost'] = (df['Cost Per Month - Basic ($)'] + df['Cost Per Month - Standard ($)'] + df['Cost Per Month - Premium ($)'])/3
df
```

```python
#Finding the average value of the basic plan
df['Avg_Basic'] = df['Cost Per Month - Basic ($)'].sum()/df['Cost Per Month - Basic ($)'].count()

#Finding the average value of standard plan
df['Avg_Standard'] = df['Cost Per Month - Standard ($)'].sum()/df['Cost Per Month - Standard ($)'].count()

#Finding the average value of premium plan
df['Avg_Premium'] = df['Cost Per Month - Premium ($)'].sum()/df['Cost Per Month - Premium ($)'].count()
```

```python
#Arranging the avg values of all the plans upto 4 characters using lambda function
df['Avg_Cost'] = df['Avg_Cost'].apply(lambda x: float(str(x)[:4]))
df['Avg_Basic'] = df['Avg_Basic'].apply(lambda x: float(str(x)[:4]))
df['Avg_Standard'] = df['Avg_Standard'].apply(lambda x: float(str(x)[:4]))
df['Avg_Premium'] = df['Avg_Premium'].apply(lambda x: float(str(x)[:4]))
df.head(10)
```

```python
#Calculating the cost difference for each plan and arranging the value upto 5 characters using lambda function
df["Basic_Diff"] = df["Cost Per Month - Basic ($)"]-df['Avg_Basic']
df["Basic_Diff"] = df["Basic_Diff"].apply(lambda x: float(str(x)[:5]))

df["Std_diff"] = df["Cost Per Month - Standard ($)"]-df['Avg_Standard']
df["Std_diff"] = df["Std_diff"].apply(lambda x: float(str(x)[:5]))

df["Prem_diff"] = df["Cost Per Month - Premium ($)"]-df['Avg_Premium']
df["Prem_diff"] = df["Prem_diff"].apply(lambda x: float(str(x)[:5]))

#Highest price of subscription
df1 = df.sort_values(by = 'Avg_Cost', ascending=False).reset_index(drop = True)[:10]
df1
```

```python
#least price of subscription
df2 = df.sort_values(by = 'Avg_Cost', ascending= True).reset_index(drop = True)[:10]
df2                                                                              
```

```python
#Plotting the country with high charges
df1.plot(kind = 'bar',x='Country',y='Avg_Cost',color='red')
plt.show()

#Plotting the country with least charge
df2.plot(kind = 'bar',x='Country',y='Avg_Cost',color='blue')
plt.show()
```
